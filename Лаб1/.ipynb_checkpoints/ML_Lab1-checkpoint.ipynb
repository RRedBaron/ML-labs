#### Importing packages

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from collections import Counter
```

#### Loading dataset to dataframe

```python
dataset = pd.read_csv("dataset.csv").rename(columns={'rate.1':'rating'})
```

#### Data preprocessing

```python
dataset.drop('rate', axis=1, inplace=True)
dataset.dropna(subset=["votes count"], inplace=True)
dataset['certificate'].fillna('M', inplace=True)
dataset['genre'].fillna('Mixed', inplace=True)
```

```python
dataset.sort_index(axis=1, ascending=False)
```

```python
figure = plt.figure(figsize=(10,5))
ax1 = figure.add_subplot(2,1,1)
ax1.set_xlabel('Ratings distribution')
n, bins, patches = ax1.hist(dataset['rating'], color='red', bins=50)
ax1.set_xticks(np.arange(0, max(dataset['rating'])+0.5, 0.5))
frequency = dict(Counter(list(dataset['genre'])))
ax2 = figure.add_subplot(2,1,2).bar(list(frequency.keys()), list(frequency.values()),width=1)
plt.xticks(rotation=90)
plt.xlabel('Games Genres')
figure.subplots_adjust(hspace=0.5, top=0.9, bottom=0.1)
figure.tight_layout()
```
